```python
import pandas as pd
import requests
import json
import os
import numpy as np
import plotly.express as px
```

```python
def get_weights_ref(concepts):
    vector_ref = concepts.copy()
    vector_ref['weight'] = 0.0
    vector_ref = vector_ref.drop('class_id,concept_id,description,mapping'.split(','), axis = 1)
    vector_ref = vector_ref.set_index('concept')
    return vector_ref

def get_weights(concepts_df, concept_selection, scale):
    weights = pd.DataFrame(index = concepts_df['concept'], data = 0.0, columns = ['weight'])
    for c, s in zip(concept_selection, scale):
        weights.loc[c] = s

    return weights

def select_dot(vectors, weights):
    return vectors.dot(weights).sort_values('weight', ascending= False)

def select_minerr(vectors, weights):
    weights = weights.loc[weights['weight'] != 0]
    vectors = vectors[weights.index]
    return pd.DataFrame((vectors - weights.values.T).sum(axis = 1).sort_values(ascending= False))
```

```python
concepts = pd.read_csv("../data/concepts.csv")
vectors = pd.read_csv('../data/vectors.csv', index_col='word')
```

```python
weights = get_weights(concepts, ['insult', 'formality'], [-1, -1])
select_dot(vectors,weights).head(10).T # targetting insults
```

```python
weights = get_weights(concepts, ['insult', 'formality'], [1,1])
select_dot(vectors,weights).head(10).T # targeting compliments (probably should've reversed polarity for this metric or changed name to compliment in retrospect)
```

```python
weights = get_weights(concepts, ['terrestrial','shelter'], [1, 1]) # targetting earthbound shelter broadly
select_dot(vectors,weights).head(10).T
```

```python
weights = get_weights(concepts, ['worth','gender', 'object', 'agency'], [1, 1, 1, 1]) # targeting queen.
select_dot(vectors,weights).head(10).T 
```

```python
weights = get_weights(concepts, ['worth','gender', 'object', 'agency'], [1, -1, 1, 1]) # targeting king.
select_dot(vectors,weights).head(10).T 
```

```python
weights = get_weights(concepts, ['artificiality','shelter', 'intelligence'], [1, 1, 1]) # targeting labratory
select_dot(vectors,weights).head(10).T
```
